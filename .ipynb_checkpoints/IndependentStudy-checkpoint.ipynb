```python
import pickle

import pandas as pd
import numpy as np
import seaborn as sns

import folium
from statsmodels.stats.weightstats import ttest_ind

import matplotlib.pyplot as plt
%matplotlib inline
from bokeh.plotting import figure
from bokeh.io import output_notebook, push_notebook, show
output_notebook()
```

The below analysis is conducted on a data set of New York City Green taxicab pick ups from the month of September 2015. I have processed the data and chosen to answer the questions illustrated below.

#### Research Questions 

1. Does Average speed differ by day of the week, weekend vs weekday?
2. Does average speed differ for airport trips vs non-airport trips
3. Does Tip percentage differ for airport trips vs. non airport trips
4. Illustrate the effect of time of day on trip distance and # of trips
5. Does Trip_distance correlate with tip_pct
6. What findings can be shown from where geopgraphically cabs pick up

#### Data Processing 

```python
df = pd.read_csv('green_tripdata_2015-09.csv')
df.describe()
```

```python
df.shape
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df.lpep_pickup_datetime)
df['Lpep_dropoff_datetime'] = pd.to_datetime(df.Lpep_dropoff_datetime)
df['pickup_hour'] = df.lpep_pickup_datetime.dt.hour 
df['dropoff_hour'] = df.Lpep_dropoff_datetime.dt.hour
```

```python
df['tip_pct'] = 100*df.Tip_amount/df.Total_amount
df = df[df.Passenger_count > 0]
```

```python
df['duration'] = pd.to_timedelta(df.Lpep_dropoff_datetime - df.lpep_pickup_datetime).dt.seconds/60
df = df[df.duration > 0]
df['duration'].describe()
```

```python
df['avg_mph'] = 60*df.Trip_distance/df.duration
df = df[df.avg_mph < 75]
df.avg_mph.describe()
```

I have chosen to remove fares where the duration is 0 and the average speed is greater than 75mph. Since in these cases something is either errant or mis reported 

```python
df['day_of_week'] = df.lpep_pickup_datetime.dt.dayofweek
df['weekend'] = np.where((df.day_of_week == 0) | (df.day_of_week == 6), 1, 0)
```

```python
df = df[df.tip_pct > 0]
df = df[df.Fare_amount > 2.50]
```

I have also removed fares with a tip of zero percent of a fare amount less than 2.50. The Minimum fare allowed in a legitimate ride is $2.50

#### Question 1

```python
ax = sns.violinplot(x='day_of_week', y='avg_mph', data=df, size=6, aspect=3)
ax.set(xlabel='day of week', ylabel='Average MPH', title='Violin Plot of Average MPH by day of week')
plt.show()
```

```python
ax = sns.violinplot(x='weekend', y='avg_mph', data=df, size=6, aspect=3)
ax.set(xlabel='Weekend?', ylabel='Average MPH', title='Violin Plot of Average MPH by on Weekend vs non-weekend')
plt.show()
```

```python
data = df.sample(10000)
x = data['pickup_hour'].values
y = data['avg_mph'].values
color = np.where(data['weekend'] == 0, '#00ff00', '#0000ff')
TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"

p = figure(tools=TOOLS)
p.scatter(x, y,
          fill_color=color, fill_alpha=0.6,
          line_color=None)

show(p)
```

```python
sample1 = df[df.weekend == 0].avg_mph.values
sample2 = df[df.weekend == 1].avg_mph.values
ttest_ind(sample1, sample2, alternative='smaller')
```

```python
print df[df.weekend == 0].avg_mph.mean()
print df[df.weekend == 1].avg_mph.mean()
```

The violin plots above show that the distributions of mile per hour by day of the week are nearly identical visually. When breaking them down by hour of the day in a scatter plot there is no visible trend. However when using a t-test to compare the average speed of trips, we have enough data to conclude that weekend trips have a larger mile per hour despite being only 1.2mph faster. Visually this is actually really hard to discern

#### Question 2

```python
df['is_airport'] = np.where((df.RateCodeID == 2) | (df.RateCodeID == 3), 1, 0)
```

```python
ax = sns.violinplot(x='is_airport', y='avg_mph', data=df, size=6, aspect=3)
ax.set(xlabel='Airport Trip?', ylabel='Average MPH', title='Violin Plot of Average MPH by for Airport/non-Airport trips')
plt.show()
```

```python
data = df.sample(10000)
x = data['pickup_hour'].values
y = data['avg_mph'].values
color = np.where(data['is_airport'] == 0, '#ff0000', '#000000')
TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"

p = figure(tools=TOOLS, x_axis_label="Pickup hour", y_axis_label='Avg MPH',
           title='Avg MPH vs. Pickup Hr. For Airport/Non-Airport Trips')
p.scatter(x, y,
          fill_color=color, fill_alpha=0.6,
          line_color=None,
          legend="Not-Airport")

show(p)
```

```python
sample1 = df[df.is_airport == 0].avg_mph.values
sample2 = df[df.is_airport == 1].avg_mph.values
ttest_ind(sample1, sample2, alternative='smaller')
```

```python
print df[df.is_airport == 0].avg_mph.mean()
print df[df.is_airport == 1].avg_mph.mean()
```

Violin Plots are excellent for displaying all of the same information you would get from a bar chart coupled with the same information you would get from a histogram. What I find particularly interesting is that we can visualize that Airport trips have a bimodal distribution for average mph. This is most likely because there are two airports contained in this data set. Newark, and JFK. Since these use a different road system one trip is likely faster on average than the other.


We can also see in the scatter plot that, while no visible trend is present, There are a much greater number of non-airport trips.

Finally, airport trips are nearly 7 miles per hour faster on average. This is because the Taxis are using the interstate system and avoiding crowded roads and red lights.

#### Question 3

```python
ax = sns.violinplot(x='is_airport', y='tip_pct', data=df, size=6, aspect=3)
ax.set(xlabel='Airport Trip?', ylabel='tip_pct', title='Violin Plot of Tip Percentage by for Airport/non-Airport trips')
plt.show()
```

```python
sample1 = df[df.is_airport == 0].tip_pct.values
sample2 = df[df.is_airport == 1].tip_pct.values
ttest_ind(sample1, sample2, alternative='two-sided')
```

```python
print df[df.is_airport == 0].tip_pct.mean()
print df[df.is_airport == 1].tip_pct.mean()
```

Airport trips have a slightly lower tip percentage at 15.8 Percent. What is interesting is that for non airport trips there are three distinct bands corresponding to 18%, 20%, and 23%. This is because the pinpad in the cabs includes these as default options for adding tips!

#### Question 4

```python
grouped = df.groupby(['pickup_hour']).agg([np.mean, np.median]).Trip_distance.reset_index()
ax = sns.lmplot(x='pickup_hour', y='mean', data=grouped, fit_reg=False, size=6, aspect=3)
ax.set(xlabel='pickup_hour', ylabel='Mean Trip Distance', title='Mean Distrance vs. Pickup Hour')
plt.show()
```

People are by far taking the longest distance trips on average during the early morning hours. The distance steadily declines throughout the day, until slowly rising again after 8pm (hr 20).

```python
grouped = df.groupby(['pickup_hour']).agg({'Trip_distance': 'count'}).reset_index()
grouped.columns = ['pickup_hour', 'count_of_trips']
ax = sns.lmplot(x='pickup_hour', y='count_of_trips', data=grouped, fit_reg=False, size=6, aspect=3)
ax.set(xlabel='pickup_hour', ylabel='Count of Trips', title='# of All Trips by Hour of Day')
plt.show()
```

We can also observe that the early morning hours are the minima of trips hailed. The reason why early morning and late night trips have longer distances is because public transport options such as the subway have slowed to a halt, the intervals between arrivals have increased, so people hail cabs for longer distance trips. Throughout the busier times of the day people only tend to take cabs for short distance trips. Clearly there is some function of cost and travel time that riders are subconsciously optimizing for.

```python
grouped = df.groupby('pickup_hour').agg({'avg_mph': 'mean'}).reset_index()
ax = sns.lmplot(x='pickup_hour', y='avg_mph', data=grouped, fit_reg=False, size=6, aspect=3)
ax.set(xlabel='Pickup Hour', ylabel='avg MPH', title='mean Avg MPH, Across the day')
```

In this case we can clearly observe that Average speed of trips is related to traffic patterns. We see the avg mph begins to drop drastically during the morning rush, and slowly rise again as the evening rush subsides.

We can see visually there doesnt seem to be a trend between trip distance and tip percentage. When plotting a histogram of trip percentage we can observe once again the patterns of pre-selected tip percentages corresponding to the pin pad in the cab. The correlation coefficient is ~-0.13, Trip distance is not a useful predictor of tip percentage. We can however see that as expected, the Tip Amount correlates to the Total Amount with a coefficient of .8!

```python
m = folium.Map([40.744318, -73.990326], tiles='Stamen Toner', zoom_start=11)
data = df.sample(1000)
points = zip(data.Pickup_latitude.values, data.Pickup_longitude.values)
for point in points:
    folium.CircleMarker([point[0], point[1]], radius=1, color='#0000ff', popup=folium.Popup("inline explicit Popup")).add_to(m)
m
```

For the selected sampling, notice that none of the cabs picked up from inside of lower manhattan. All of the fares originate in the outer boroughs of Manhattan. This is because the NYC green cabs are only allowed to pick up in the outer boroughs. Also of note they can only pick up from the airports when arranged in advance. The reason for this is because data has illustrated historically that cabs would not pick up in the outer boroughs or lower income areas so the NYC taxicab commission opened up the market for green cabs that can only pick up in these designated locations
